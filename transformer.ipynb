```python
# Open the file 'data.txt', read all the text inside, and save it to a variable called 'text'
text = open("/content/sample_data/data.txt").read()

# Get unique characters using set(), Turn them into a list, Sort them alphabetically
# This creates our alphabet (e.g., [' ', '!', 'a', 'b', 'c'...])
chars = sorted(list(set(text)))

# Create a dictionary that maps each character (string) to a unique number (integer)
# Example: {'a': 0, 'b': 1}
stoi = {ch: i for i, ch in enumerate(chars)}

# Create a dictionary that does the reverse: maps the number back to the character
# Example: {0: 'a', 1: 'b'}
itos = {i: ch for ch, i in stoi.items()}

# Store the total count of unique characters (the size of our "alphabet")
vocab_size = len(chars)
```
